### import packages 

```python
import pandas as pd 
```

### import master file 

```python
df_master = pd.read_csv("/Users/baselhussein/Projects/pps_isls_2022/data/cleanFiles/df_master.csv")
```

```python
df_master.head(3)
```

### FLOAT middle summary

```python
### select FLOAT middle

df_float_middle = pd.DataFrame(df_master.loc[(df_master['level'] == "float") & (df_master['group'] == "middle")])
df_float_middle.head(3)
```

```python
### get Problem solving moves

df_float_middle_summary = df_float_middle[["ï»¿filename", "Problem_Solving_Move"]]
df_float_middle_summary
```

```python
df_float_middle_move = pd.DataFrame(df_float_middle["Problem_Solving_Move"])
df_float_middle_move.head()
```

```python
### get unique moves

categories_moves = df_float_middle_move['Problem_Solving_Move'].unique()
categories_moves
```

```python
### get matrix count of moves

df_float_middle_move = pd.get_dummies(df_float_middle_move.Problem_Solving_Move).reindex(columns=categories_moves, fill_value=0)
```

```python
df_float_middle_move
```

```python
### combine counts with moves

df_float_middle_summary_moves = pd.concat([df_float_middle_summary, df_float_middle_move], axis=1)
df_float_middle_summary_moves
```

```python
### drop irrelevant values

df_float_middle_summary_moves.drop('Problem_Solving_Move', axis=1, inplace=True)
```

```python
df_float_middle_summary_moves
```

```python
### get count by player

byPlayer = df_float_middle_summary_moves.groupby('ï»¿filename')
```

```python
df_float_middle_sum = pd.DataFrame(byPlayer.sum())
df_float_middle_sum
```

```python
### get mean moves by player

df_float_middle_sum.mean()
```

```python
### get std moves between players

df_float_middle_sum.std()
```
